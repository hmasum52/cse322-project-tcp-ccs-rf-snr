```python
import pandas as pd
import matplotlib.pyplot as plt
import os
data_file = "wired.csv"
modified_data_file = "wired_modified.csv"
```

```python
def plot_graph(x_list, y_list, y_list_m,_marker="o", x_label="", y_label=""):
    plt.plot(x_list, y_list,marker=_marker, color="blue")
    plt.plot(x_list, y_list_m,marker=_marker, color="red")
    plt.scatter(x_list, y_list, color="red")
    plt.scatter(x_list, y_list_m, color="blue")
    plt.xlabel(xlabel=x_label)
    plt.ylabel(ylabel=y_label)
    plt.legend([y_label, y_label+" modified"])
    # show grid
    plt.grid()
    plt.title(f"{y_label} vs {x_label}")
    if(not os.path.exists("graph")):
        os.mkdir("graph")
    plt.savefig(f"graph/{x_label}_vs_{y_label}.png")
    plt.show()

def parse_and_plot(data, data_m, x_col, y_col):
    params = data.columns.values.tolist()
    # x_list data is the xth column of the data
    x_list = data[params[x_col]].values.tolist()
    # y_list data is the yth column of the data
    y_list = data[params[y_col]].values.tolist()
    y_list_m = data_m[params[y_col]].values.tolist()
    print(params[x_col])
    print(y_list)
    print(y_list_m)
    plot_graph(x_list, y_list, y_list_m, x_label=params[x_col], y_label=params[y_col])
```

```python
# variable node list
var_node_list = pd.read_csv(data_file, skiprows=2, nrows=5)
print(var_node_list)
var_node_list_modified = pd.read_csv(modified_data_file, skiprows=2, nrows=5)
print(var_node_list_modified)
```

```python
params = var_node_list.columns.values.tolist()
print(params)

areas = var_node_list[params[0]].values.tolist()
print(areas)

# Area vs [Throughput, Average delay, Delivery ratio, Drop ratio]
for i in range(1, len(params)):
    parse_and_plot(var_node_list, var_node_list_modified, 0, i)
```

```python
# variable flow list
flow_list = pd.read_csv(data_file, skiprows=9, nrows=5)
print(flow_list)
flow_list_modified = pd.read_csv(modified_data_file, skiprows=9, nrows=5)
print(flow_list_modified)
```

```python
params = flow_list.columns.values.tolist()
print(params)

# Number of nodes vs [Throughput, Average delay, Delivery ratio, Drop ratio]
for i in range(1, len(params)):
    parse_and_plot(flow_list, flow_list_modified, 0, i)
```

```python
# variable packet rate list
pkt_rate_list = pd.read_csv(data_file, skiprows=16, nrows=5)
print(pkt_rate_list)
pkt_rate_list_modified = pd.read_csv(modified_data_file, skiprows=16, nrows=5)
print(pkt_rate_list_modified)
```

```python
params = pkt_rate_list.columns.values.tolist()
print(params)

# Number of node flow vs 
# [Throughput, Average delay, Delivery ratio, Drop ratio]
for i in range(1, len(params)):
    parse_and_plot(pkt_rate_list, pkt_rate_list_modified, 0, i)
```
